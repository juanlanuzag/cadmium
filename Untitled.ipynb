```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import copy
from matplotlib import animation
from IPython.display import HTML
sns.set()
```

```python
def get_position_i(row):
    return int(row['cellname'].split('_')[-2])

def get_position_j(row):
    return int(row['cellname'].split('_')[-1])


def get_data():
    data = pd.read_csv('output.txt', header=None, names=['Time', 'cellname', 'State'])
    data['i'] = data.apply (lambda row: get_position_i(row), axis=1)
    data['j'] = data.apply (lambda row: get_position_j(row), axis=1)
    return data
```

```python
def get_width(df):
    return max(df['i']) +1

def get_frames(df):
    frames = []
    times = set(data['Time'])
    
    curr_frame = [None] * get_width(df)
    for instant in times:
        #display(data[data['Time'] == instant])
        for idx, row in data[data['Time'] == instant].iterrows():
            curr_frame[int(row.i)] = int(row.State)
        frames.append(copy.deepcopy(curr_frame))
    return frames
```

```python
data=get_data()
fig = plt.gcf()
fig.set_size_inches(100, 10.5)
sns.heatmap(get_frames(data), square=True, cbar=False, linewidths=0, linecolor='black',cmap=['white', 'black'])
plt.show()
```

```python
frames= get_frames(data)
fig = plt.gcf()
fig.set_size_inches(40, 20)

def animate(i):
    sns.heatmap(frames[:i+1], square=True, yticklabels=range(40), cbar=False, linewidths=0, linecolor='black',cmap=['white', 'black'])

def init():
    animate(0)

    
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=40, repeat = True, interval=250)
HTML(anim.to_html5_video())
```
